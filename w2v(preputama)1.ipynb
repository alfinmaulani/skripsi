```python
import numpy as np
import pandas as pd
import io
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

url = 'https://raw.githubusercontent.com/alfinmaulani/skripsi/master/dataisear(dokumen).csv'
df = pd.read_csv(url)
# df.iloc[:,1]
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
from gensim.test.utils import datapath

# sentences = LineSentence(url)

sentences = [word_tokenize(x) for x in df.dokumen] 
sentences[:2]

# id_w2v = Word2vec Word2vec (sentences, size=100, workers=multiprocessing.cpu_count() 1)
# model.save("word2vec.model")
```

```python
w2v = Word2Vec(sentences, window=5, min_count=1, workers=4, cbow_mean=0)
```

```python
cek=w2v.wv
words=w2v.wv.vocab
# words
cek.vector_size
```

```python
w1=w2v.wv["moment"]
print(w1)
```

```python
w2v.save('w2v(preputama).model')
```
