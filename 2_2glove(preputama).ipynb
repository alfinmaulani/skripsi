```python
import numpy as np
import pandas as pd
import io
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

url1 = 'https://raw.githubusercontent.com/alfinmaulani/skripsi/master/dataisear(preputama).csv'
url2 = 'tba'
df = pd.read_csv(url1)
# df.iloc[:,1]
```

```python
# for x in train.dokumen:
#   print(x)
sentences = [word_tokenize(x) for x in df.dokumen] 
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer=Tokenizer(num_words=num_words)
tokenizer.fit_on_texts(df.dokumen)
```

```python
train_sequences=tokenizer.texts_to_sequences(df.dokumen)
```

```python
train_sequences
```

```python
train_padded= pad_sequences(train_sequences, maxlen=178, padding='post', truncating='post')
```
