convert string to list

```python
def convert_to_list(tem):
  temp=[]
  for senten in tem:
    temp.append(ast.literal_eval(senten))
  return temp
```

```python
sent=convert_to_list(sent)
```

word2vec

```python
from gensim.models import Word2Vec
```

```python
w2v = Word2Vec(sent, window=5, min_count=1, sg=1, iter=10)
```
